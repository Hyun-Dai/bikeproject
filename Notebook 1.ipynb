## Problem Statement

How do bike usage trends compare to usage trends for other vehicles (Buses/Coaches, Cars/Vans/Taxis, Motorcycles, and Trains/Rails) in the United States and United Kingdom? Additionally, how do these usage trends impact the amount of carbon dioxide emitted in the United States?

## Assumptions

- No new significant technology that heavily affects transportation types or carbon emission rates in the future
  - Justification: We must assume that the transportation remains relatively constant in order to relate changes in various factors specifically to greater use of e-bikes.
  
- Metrics will use average yearly rates of carbon emmissions per transportation type mentioned in the problem statement to calculate carbon dioxide emissions from each transportation type with consistency.
      1. Passenger car: 4600 kg. per year per car
      2. Motorcycle: 478.7 kg. per year per motorcycle
      3. Person: 82.5 kg. per year per person
      4. Biking: 70 kg. per year per bike
      5. Mass Transport/Rail: 117 tons per line per year



## Computing

```
# Import libraries to process and graph data
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
# Extract Data from CSV Files
# All data is chosen from 1990 to 2020
# UK E-Bike Sales Data Extracted from gov.uk (https://www.gov.uk/government/statistical-data-sets/tsgb01-modal-comparisons#passenger-transport) (Passenger journeys on public transport vehicles from 1950)
uk_data2020 = pd.read_csv("UK - Sheet1(1).csv")
# US E-Bike Sales Data Extracted from Bureau of Transportation (https://www.bts.gov/content/us-passenger-miles)
us_data2020 = pd.read_csv("US - Sheet1(1).csv")
```

```
# UK Data categories used for polynomial Regression 
variables = ['Buses and coaches', 'Cars, vans and taxis', 'Motorcycles','Bicycles', 'Rail']

# For loop to go through all categories
for var in variables:
    plt.figure() # Creating a rectangle (figure) for each plot
    fig, ax = plt.subplots()
    # These years were chosen because we balance the need for past data to make predictions with the
    # fact that the further in the past data is, the worse predictor of the future it becomes
    xlim = [1990, 2030]
    ax.set_xlim(xlim)
    # Add ylim to limit y-axis to prevent the graph going negative
        # ensures the number of people is nonnegative, in case it reaches negative over the error bounds;
        # commented to see where it goes; uncomment for specific purposes
        # could be necessary, useful for very specific purposes
    # ylim = [0, 50]
    # ax.set_ylim(ylim)
    sns.regplot(x='Year', y=var, data=uk_data2020, order=2, truncate=False).set(title=f'UK Regression plot of Year and {var}');
    # Add units for clarification
    ax.set_ylabel(f"{var} (in billions of km)" , size = 12)
```

```
# US Data Polynomial Regression
variables = ['Passenger vehicles (trucks, cars, motorcycles, etc.)', 
'Mass Transportation (trains, buses, ferries, etc.)', 'Bicycles','Walking']
# variables = ['Walking']
for var in variables:
    plt.figure() # Creating a rectangle (figure) for each plot
    fig, ax = plt.subplots()
    xlim = [1990, 2030]
    ax.set_xlim(xlim)
    # Add ylim to limit y-axis to prevent the graph going negative
    # ylim = [0, 50]
    # ax.set_ylim(ylim)
    sns.regplot(x='Year', y=var, data=us_data2020, order=2, truncate=False).set(title=f'US Regression plot of Year and {var}');
    # Add units for clarification
    ax.set_ylabel(f"{var} (in billions of miles)" , size = 6)
```

## Analysis

The kgCOe/vehicle is given is the Assumptions area.

United States Carbon Emissions Impacts due to the above predictions:

- Passenger vehicles: (38.6 thousand vehicles) * (4600 kgCOe/vehicle) = 177,560 less metric tons of carbon emissions. 

- Bikes: (3.32 thousand bikes) * (70 kgCOe/bike) = 232.4 additional metric tons of carbon emissions. 

- Mass transport/Rail: (6.04 thousand passengers) * (117 tonsCOe/mass transport) = 706,880 less metric tons of carbon emissions. 

- Walking: (0.92 thousand pedestrians) * (82.5 kgCOe/person) = 75.9 less metric tons of carbon emissions. 

Overall, the US will see a reduction of 884,283.5 metric tons of carbon emissions due to the growth of e-bike sales.

## Model Strengths and Weaknesses

Strengths: Utilizing a polynomial regression model allows for flexibility to account for much of the dataset, offering a concise relationship between the independent and dependent variable. Furthermore, the usage of a medium degree order of polynomials (that of 2) ensures the regression model properly covers the majority of the data without overfitting.


Weaknesses: The primary weakness of the polynomial regression model lies in its poor handling of outliers. Data points that originate from the post-COVID era have been significantly affected by a severe decrease in transportation. While we made sure to include this data nonetheless in the regression model to include a full picture, outliers provide noise. In the end, this may not be the fault of the polynomial regression model itself; predicting the effects of transportation in the still young era of COVID is difficult.
